# Agregando una nueva columna con un array

El dataframe tiene 1000 filas, asi que se necesita un array de 1000 elementos

```python
# Importar numpy
import numpy as np
```

```python
# Creamos un array de 1000 elementos
languajeScore = np.arange(0, 1000)
```

```python
# Tamaño del array
len(languajeScore)
```

```python
# Agregar una nueva columna al dataframe con un array
dfExamenes['languaje score'] =languajeScore
dfExamenes
```

```python
# Creando valores aleatorios entre 1 y 100
intLanguageScore = np.random.randint(1, 100, size = 1000)
```

```python
# Maximo y minimo
print(min(intLanguageScore))
print(max(intLanguageScore))
```

```python
# Agregando los nuevos valores a la columna del dataframe con un array
dfExamenes['languaje score'] = intLanguageScore
dfExamenes
```

```python
# Creando numeros aleatorios flotantes entre 1 y 100
np.random.uniform(1, 100, size = 1000)
```
