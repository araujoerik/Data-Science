```python
import pandas as pd
```

```python
# Leyendo el archivo CSV
dfExamenes = pd.read_csv('DesempenioEstudiantes.csv')
dfExamenes.head()
```

# Operaciones

### Operaciones en columnas

```python
# Selecciona una columna y calcula la suma total
dfExamenes['math score'].sum()
```

```python
# Contar, promedio, desviación estandar, maximo y minimo
print("El numero de elementos es: " + str(dfExamenes['math score'].count()))
print("El promedio de los elementos es: " + str(dfExamenes['math score'].mean()))
print("La desviacion estandar de los elementos es: " + str(dfExamenes['math score'].std()))
print("El valor maximo de los elementos es: " + str(dfExamenes['math score'].max()))
print("El valor minimo de los elementos es: " + str(dfExamenes['math score'].min()))
```

```python
# Se puede simplificar con el metodo .describe()
dfExamenes.describe()
```

### Operaciones en filas

```python
# Calculando la suma en una fila
dfExamenes['math score'] + dfExamenes['reading score'] + dfExamenes['writing score']
```

```python
# Calculando el promedio y asignar los resultados a una nueva columna
dfExamenes['average'] = dfExamenes['math score'] + dfExamenes['reading score'] + dfExamenes['writing score']
dfExamenes['average'] = dfExamenes['average'] / 3
```
